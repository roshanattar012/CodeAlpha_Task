```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# For better graph style
sns.set_style("darkgrid")

# Show plots inside notebook
%matplotlib inline
```

```python
df = pd.read_csv("C:\\Users\\rosha\\OneDrive\\Desktop\\CodeAlpha_EDA_Project\\data\\Datafiniti_Amazon_Consumer_Reviews_of_Amazon_Products.csv")

# Display first 5 rows
df.head()
```

```python
# Shape of dataset
print("Rows and Columns:", df.shape)

# Column names
print("\nColumns:")
print(df.columns)

# Dataset information
print("\nDataset Info:")
print(df.info())
```

```python
print(df.isnull().sum())
```

```python
df = df[['name', 'reviews.rating', 'reviews.text']]

# Rename columns
df.columns = ['Product', 'Rating', 'Review']

df.head()
```

```python
df.dropna(inplace=True)

print(df.isnull().sum())
```

```python
print(df.describe())
```

```python
print("Total Unique Products:")
print(df['Product'].nunique())
```

```python
plt.figure(figsize=(8,5))

sns.countplot(x='Rating', data=df)

plt.title("Distribution of Product Ratings")

plt.xlabel("Ratings")

plt.ylabel("Number of Reviews")

plt.show()
```

```python
top_products = df['Product'].value_counts().head(10)

plt.figure(figsize=(12,6))

top_products.plot(kind='bar')

plt.title("Top 10 Most Reviewed Products")

plt.xlabel("Products")

plt.ylabel("Review Count")

plt.xticks(rotation=75)

plt.show()
```

```python
# Create review length column
df['Review_Length'] = df['Review'].apply(len)

plt.figure(figsize=(10,5))

sns.histplot(df['Review_Length'], bins=30)

plt.title("Review Length Distribution")

plt.xlabel("Review Length")

plt.show()
```

```python
plt.figure(figsize=(6,4))

sns.heatmap(df[['Rating', 'Review_Length']].corr(),
            annot=True,
            cmap='coolwarm')

plt.title("Correlation Heatmap")

plt.show()
```

# Key Insights

1. Most customers gave high ratings.
2. Certain products received significantly more reviews.
3. Longer reviews often contain detailed customer experiences.
4. Review patterns help understand customer satisfaction.
